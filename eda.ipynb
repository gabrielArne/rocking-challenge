```python
df = pd.concat([disney_df, netflix_df], ignore_index=True, sort=False)
df.head()
```

```python
#veo si algun id de show_id no comienza con la letra "s"
df[~df["show_id"].str.startswith("s")]
```

```python
# la columna duration debe ser abierta en duracion y tipo 

df["duration"].str.split(" ", expand=True).iloc[:, 1].unique()
```

```python
df[["duration_int", "duration_type"]] = df["duration"].str.split(" ", expand=True)
df.drop(columns=["duration"], inplace=True)
```

```python
#guardo el csv raw combinado
df.to_csv("./data/raw/combined.csv", index=False)
```
